```python
import numpy as np

def damped_sine_wave(x, amplitude=1, frequency=1, damping=0.1):
    """
    Computes a damped sine wave function.
    
    Parameters:
        x (array-like): Input values.
        amplitude (float): Amplitude of the wave.
        frequency (float): Frequency of the sine wave.
        damping (float): Damping coefficient.
    
    Returns:
        array-like: Computed y-values.
    """
    return amplitude * np.exp(-damping * x) * np.sin(2 * np.pi * frequency * x)
```

```python
import matplotlib.pyplot as plt

# Generate x values
x_values = np.linspace(0, 10, 500)  # From 0 to 10 with 500 points

# Compute y values
y_values = damped_sine_wave(x_values, amplitude=1, frequency=2, damping=0.2)

# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='Damped Sine Wave', color='b')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Damped Sine Wave')
plt.legend()
plt.grid(True)
plt.show()
```
